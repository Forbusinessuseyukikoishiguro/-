```python
# 小学生向けプログラミングスクール講座企画ツール
# Google Colab用 - GPT APIを使用

# 必要なライブラリのインストール
!pip install openai

import openai
import json
from datetime import datetime
import pandas as pd

# OpenAI APIキーの設定
# 注意: 実際のAPIキーを入力してください
OPENAI_API_KEY = "sk-ほげほげあなたのAPIキー"
openai.api_key = OPENAI_API_KEY

class ProgrammingCourseTemplate:
    """小学生向けプログラミング講座のテンプレートクラス"""

    def __init__(self):
        self.template = {
            "基本情報": {
                "講座名": "",
                "対象年齢": "",
                "受講人数": "",
                "総時間数": "",
                "実施期間": ""
            },
            "学習目標": {
                "主目標": "",
                "副目標": [],
                "期待する成果": []
            },
            "使用技術・ツール": {
                "プログラミング言語": "",
                "開発環境": "",
                "必要な機材": [],
                "追加ツール": []
            },
            "カリキュラム構成": [],
            "評価方法": {
                "評価基準": [],
                "成果物": "",
                "発表形式": ""
            },
            "準備・注意事項": {
                "事前準備": [],
                "安全対策": [],
                "保護者向け説明": ""
            }
        }

    def create_course_plan(self):
        """対話形式で講座企画を作成"""
        print("=== 小学生向けプログラミング講座企画作成 ===\n")

        # 基本情報の入力
        print("【基本情報】")
        self.template["基本情報"]["講座名"] = input("講座名: ")
        self.template["基本情報"]["対象年齢"] = input("対象年齢（例: 小学3-6年生）: ")
        self.template["基本情報"]["受講人数"] = input("受講人数: ")
        self.template["基本情報"]["総時間数"] = input("総時間数（例: 10時間）: ")
        self.template["基本情報"]["実施期間"] = input("実施期間（例: 2ヶ月間、週1回）: ")

        # 学習目標の入力
        print("\n【学習目標】")
        self.template["学習目標"]["主目標"] = input("主な学習目標: ")

        print("副目標を入力してください（終了するには空白でEnter）:")
        sub_goals = []
        while True:
            goal = input("副目標: ")
            if goal.strip() == "":
                break
            sub_goals.append(goal)
        self.template["学習目標"]["副目標"] = sub_goals

        print("期待する成果を入力してください（終了するには空白でEnter）:")
        outcomes = []
        while True:
            outcome = input("期待する成果: ")
            if outcome.strip() == "":
                break
            outcomes.append(outcome)
        self.template["学習目標"]["期待する成果"] = outcomes

        # 使用技術・ツールの入力
        print("\n【使用技術・ツール】")
        self.template["使用技術・ツール"]["プログラミング言語"] = input("プログラミング言語（例: Scratch, Python等）: ")
        self.template["使用技術・ツール"]["開発環境"] = input("開発環境: ")

        print("必要な機材を入力してください（終了するには空白でEnter）:")
        equipment = []
        while True:
            item = input("機材: ")
            if item.strip() == "":
                break
            equipment.append(item)
        self.template["使用技術・ツール"]["必要な機材"] = equipment

        # カリキュラム構成の入力
        print("\n【カリキュラム構成】")
        print("各回の内容を入力してください（終了するには空白でEnter）:")
        curriculum = []
        session_num = 1
        while True:
            content = input(f"第{session_num}回の内容: ")
            if content.strip() == "":
                break
            curriculum.append({
                "回数": session_num,
                "内容": content,
                "時間": input("  時間: "),
                "目標": input("  この回の目標: ")
            })
            session_num += 1
        self.template["カリキュラム構成"] = curriculum

        # 評価方法の入力
        print("\n【評価方法】")
        print("評価基準を入力してください（終了するには空白でEnter）:")
        criteria = []
        while True:
            criterion = input("評価基準: ")
            if criterion.strip() == "":
                break
            criteria.append(criterion)
        self.template["評価方法"]["評価基準"] = criteria
        self.template["評価方法"]["成果物"] = input("最終成果物: ")
        self.template["評価方法"]["発表形式"] = input("発表形式: ")

        return self.template

    def display_template(self, course_plan):
        """企画内容を見やすく表示"""
        print("\n" + "="*50)
        print("講座企画書")
        print("="*50)

        for section, content in course_plan.items():
            print(f"\n【{section}】")
            if isinstance(content, dict):
                for key, value in content.items():
                    if isinstance(value, list):
                        print(f"  {key}:")
                        for item in value:
                            if isinstance(item, dict):
                                print(f"    第{item['回数']}回: {item['内容']}")
                                print(f"      時間: {item['時間']}")
                                print(f"      目標: {item['目標']}")
                            else:
                                print(f"    - {item}")
                    else:
                        print(f"  {key}: {value}")
            elif isinstance(content, list):
                for item in content:
                    if isinstance(item, dict):
                        print(f"  第{item['回数']}回: {item['内容']}")
                        print(f"    時間: {item['時間']}")
                        print(f"    目標: {item['目標']}")
                    else:
                        print(f"  - {item}")

class AIReviewer:
    """GPT APIを使用したAIレビュアークラス"""

    def __init__(self, api_key):
        self.client = openai.OpenAI(api_key=api_key)

    def review_course_plan(self, course_plan):
        """講座企画をAIがレビュー"""

        # 企画内容をテキスト形式に変換
        plan_text = self._format_plan_for_ai(course_plan)

        prompt = f"""
以下は小学生向けプログラミングスクールの講座企画書です。
教育の専門家として、この企画を詳細にレビューし、改善提案をしてください。

【レビュー観点】
1. 小学生の発達段階に適しているか
2. 学習目標が明確で達成可能か
3. カリキュラムの構成が論理的か
4. 時間配分が適切か
5. 評価方法が適切か
6. 安全面や配慮事項
7. 保護者や学校との連携

【企画書】
{plan_text}

【レビュー結果】
以下の形式で回答してください：

## 総合評価
（5段階評価と総合コメント）

## 良い点
（具体的な評価ポイント）

## 改善提案
（具体的な改善案と理由）

## 追加検討事項
（検討すべき追加要素）

## 実施時の注意点
（実際に講座を行う際の注意事項）
"""

        try:
            response = self.client.chat.completions.create(
                model="gpt-4",
                messages=[
                    {"role": "system", "content": "あなたは小学生向けプログラミング教育の専門家です。教育効果を最大化し、子どもたちが楽しく学べる講座設計についてアドバイスしてください。"},
                    {"role": "user", "content": prompt}
                ],
                temperature=0.7
            )
            return response.choices[0].message.content
        except Exception as e:
            return f"AIレビューでエラーが発生しました: {str(e)}"

    def get_improvement_suggestions(self, course_plan, specific_concern=""):
        """特定の懸念事項に対する改善提案を取得"""

        plan_text = self._format_plan_for_ai(course_plan)

        if specific_concern:
            concern_text = f"\n特に以下の点について詳しくアドバイスしてください：\n{specific_concern}"
        else:
            concern_text = ""

        prompt = f"""
以下の小学生向けプログラミング講座企画について、具体的な改善提案をしてください。
{concern_text}

【企画書】
{plan_text}

実践的で具体的なアドバイスをお願いします。
"""

        try:
            response = self.client.chat.completions.create(
                model="gpt-4",
                messages=[
                    {"role": "system", "content": "あなたは経験豊富なプログラミング教育者です。実践的で具体的なアドバイスを提供してください。"},
                    {"role": "user", "content": prompt}
                ],
                temperature=0.8
            )
            return response.choices[0].message.content
        except Exception as e:
            return f"改善提案の取得でエラーが発生しました: {str(e)}"

    def _format_plan_for_ai(self, course_plan):
        """企画内容をAI用のテキスト形式に変換"""
        formatted_text = ""

        for section, content in course_plan.items():
            formatted_text += f"\n【{section}】\n"

            if isinstance(content, dict):
                for key, value in content.items():
                    if isinstance(value, list):
                        formatted_text += f"{key}:\n"
                        for item in value:
                            if isinstance(item, dict):
                                formatted_text += f"  第{item['回数']}回: {item['内容']} (時間: {item['時間']}, 目標: {item['目標']})\n"
                            else:
                                formatted_text += f"  - {item}\n"
                    else:
                        formatted_text += f"{key}: {value}\n"
            elif isinstance(content, list):
                for item in content:
                    if isinstance(item, dict):
                        formatted_text += f"第{item['回数']}回: {item['内容']} (時間: {item['時間']}, 目標: {item['目標']})\n"
                    else:
                        formatted_text += f"- {item}\n"

        return formatted_text

# メイン実行部分
def main():
    print("小学生向けプログラミングスクール講座企画ツール")
    print("=" * 50)

    # APIキーの確認
    if OPENAI_API_KEY == "your-api-key-here":
        print("⚠️ OpenAI APIキーを設定してください")
        print("OPENAI_API_KEY = 'your-actual-api-key' に変更してください")
        return

    # 講座企画の作成
    template = ProgrammingCourseTemplate()

    print("\n1. 新しい講座企画を作成")
    print("2. サンプル企画でテスト")
    choice = input("選択してください (1 or 2): ")

    if choice == "1":
        course_plan = template.create_course_plan()
    else:
        # サンプル企画
        course_plan = {
            "基本情報": {
                "講座名": "はじめてのScratchプログラミング",
                "対象年齢": "小学3-5年生",
                "受講人数": "8名",
                "総時間数": "8時間",
                "実施期間": "4週間（週2時間）"
            },
            "学習目標": {
                "主目標": "プログラミングの基本概念を理解し、簡単なゲームを作れるようになる",
                "副目標": ["論理的思考力の向上", "創造性の発揮", "問題解決能力の育成"],
                "期待する成果": ["オリジナルゲームの完成", "プログラミングへの興味関心の向上"]
            },
            "使用技術・ツール": {
                "プログラミング言語": "Scratch 3.0",
                "開発環境": "Scratch公式サイト",
                "必要な機材": ["パソコン", "マウス", "インターネット接続"],
                "追加ツール": []
            },
            "カリキュラム構成": [
                {"回数": 1, "内容": "Scratchの基本操作とキャラクター移動", "時間": "2時間", "目標": "基本操作に慣れる"},
                {"回数": 2, "内容": "条件分岐と繰り返し処理", "時間": "2時間", "目標": "プログラミングの基本概念を理解"},
                {"回数": 3, "内容": "音と効果の追加", "時間": "2時間", "目標": "プログラムを豊かにする方法を学ぶ"},
                {"回数": 4, "内容": "オリジナルゲーム制作と発表", "時間": "2時間", "目標": "学んだことを活用して作品を完成させる"}
            ],
            "評価方法": {
                "評価基準": ["積極的な参加", "作品の完成度", "創造性", "発表での説明力"],
                "成果物": "オリジナルScratchゲーム",
                "発表形式": "各自5分間のプレゼンテーション"
            },
            "準備・注意事項": {
                "事前準備": ["Scratchアカウントの作成", "パソコンの動作確認"],
                "安全対策": ["適切な休憩時間の確保", "眼精疲労への配慮"],
                "保護者向け説明": "家庭でも継続して学習できるよう、Scratchの使い方を説明資料で提供"
            }
        }

    # 企画書の表示
    template.display_template(course_plan)

    # AIレビューの実行
    print("\n" + "="*50)
    print("AIレビューを開始します...")
    print("="*50)

    ai_reviewer = AIReviewer(OPENAI_API_KEY)

    # 基本レビュー
    review_result = ai_reviewer.review_course_plan(course_plan)
    print("\n【AIレビュー結果】")
    print(review_result)

    # 追加の相談機能
    while True:
        print("\n" + "-"*30)
        print("追加で相談したい点はありますか？")
        print("（例: 時間が足りないかもしれません、難易度が心配です、など）")
        print("終了する場合は 'quit' と入力してください")

        concern = input("\n相談内容: ")
        if concern.lower() == 'quit':
            break

        if concern.strip():
            print("\nAIが回答を生成中...")
            suggestion = ai_reviewer.get_improvement_suggestions(course_plan, concern)
            print("\n【AIからの回答】")
            print(suggestion)

    print("\n講座企画作成とレビューが完了しました！")
    print("素晴らしい講座になりますように！")

# 実行
if __name__ == "__main__":
    main()
```

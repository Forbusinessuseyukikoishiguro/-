````python
#ğŸ¾ Colabã«ã‚ƒã‚“ã“ãƒãƒ£ãƒƒãƒˆãƒœãƒƒãƒˆï¼ˆGPT-4oéŸ³å£°ä»˜ãï¼‰20250429

#ã€Œã«ã‚ƒã‚“ã“ãƒãƒ£ãƒƒãƒˆï¼ˆéŸ³å£°å¯¾å¿œğŸ¾ï¼‰ã€Colabã§ã®ä½¿ã„æ–¹

---

## ğŸ± ã«ã‚ƒã‚“ã“ãƒãƒ£ãƒƒãƒˆã®ä½¿ã„æ–¹ï¼ˆGoogle Colabç‰ˆï¼‰

ã“ã®ã«ã‚ƒã‚“ã“ãƒãƒ£ãƒƒãƒˆã§ã¯ã€**ãƒ†ã‚­ã‚¹ãƒˆã§è³ªå•ã‚’å…¥åŠ›ã™ã‚‹ã¨ã€ã‹ã‚ã„ã„ã«ã‚ƒã‚“ã“ãŒå£°ã¨æ–‡å­—ã§ãŠè¿”äº‹ã—ã¦ãã‚Œã‚‹ã«ã‚ƒï¼**
ä»¥ä¸‹ã®æ‰‹é †ã§ã‚«ãƒ³ã‚¿ãƒ³ã«ä½¿ãˆã‚‹ã«ã‚ƒï¼

---

### ğŸª„ ã‚¹ãƒ†ãƒƒãƒ—â‘ ï¼šOpenAI APIã‚­ãƒ¼ã‚’ç”¨æ„ã™ã‚‹ã«ã‚ƒï¼

[https://platform.openai.com/account/api-keys](https://platform.openai.com/account/api-keys)
ä¸Šè¨˜ãƒªãƒ³ã‚¯ã‹ã‚‰ **è‡ªåˆ†ã®APIã‚­ãƒ¼**ï¼ˆ`sk-xxxxxxxxxxxxxxxxx`ã®å½¢å¼ï¼‰ã‚’ã‚³ãƒ”ãƒ¼ã—ã¦ãŠãã«ã‚ƒï¼

---

### ğŸ¾ ã‚¹ãƒ†ãƒƒãƒ—â‘¡ï¼šColabã§ã‚³ãƒ¼ãƒ‰ã‚’å®Ÿè¡Œã™ã‚‹ã«ã‚ƒï¼

1. Google Colabã‚’é–‹ãã«ã‚ƒã€‚
2. ä»¥ä¸‹ã®ã‚³ãƒ¼ãƒ‰ã‚’ **ã‚³ãƒ”ãƒš**ã—ã¦ã€1ã‚»ãƒ«ãšã¤å®Ÿè¡Œã™ã‚‹ã«ã‚ƒã€‚
3. æœ€åˆã®ã»ã†ã§ `APIã‚­ãƒ¼ã‚’å…¥åŠ›ã—ã¦ãã ã•ã„ã€‚` ã¨ã„ã†å ´æ‰€ãŒã‚ã‚‹ã‹ã‚‰ã€è‡ªåˆ†ã®APIã‚­ãƒ¼ã«æ›¸ãæ›ãˆã‚‹ã«ã‚ƒï¼

```python
# å¿…è¦ãªãƒ©ã‚¤ãƒ–ãƒ©ãƒªã‚’ã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«ã™ã‚‹ã«ã‚ƒï¼
!pip install openai

# openaiãƒ©ã‚¤ãƒ–ãƒ©ãƒªã‚’èª­ã¿è¾¼ã‚€ã«ã‚ƒ
import os
import base64
from openai import OpenAI
from IPython.display import Audio, display

# APIã‚­ãƒ¼ã‚’ã“ã“ã«ã‚»ãƒƒãƒˆã™ã‚‹ã«ã‚ƒï¼ï¼ˆå¿…ãšã”è‡ªèº«ã®ã‚­ãƒ¼ã«æ›¸ãæ›ãˆã¦ã«ã‚ƒï¼ï¼‰
os.environ["OPENAI_API_KEY"] = "sk-xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx"

# OpenAIã‚¯ãƒ©ã‚¤ã‚¢ãƒ³ãƒˆã‚’ä½œã‚‹ã«ã‚ƒ
client = OpenAI()

# ğŸ¾ ã«ã‚ƒã‚“ã“ãƒãƒ£ãƒƒãƒˆã‚¹ã‚¿ãƒ¼ãƒˆé–¢æ•°ã ã«ã‚ƒï¼
def nyanko_chat():
    user_text = input("ã«ã‚ƒã‚“ã“ã«èããŸã„ã“ã¨ã‚’å…¥åŠ›ã—ã¦ã«ã‚ƒ: ")

    completion = client.chat.completions.create(
        model="gpt-4o-audio-preview",
        modalities=["text", "audio"],
        audio={"voice": "nova", "format": "wav"},
        messages=[
            {
                "role": "system",
                "content": "ã‚ãªãŸã¯ã‹ã‚ã„ã„ã«ã‚ƒã‚“ã“ã§ã™ã€‚èªå°¾ã«ã€Œã«ã‚ƒã€ã‚’ã¤ã‘ã¦å„ªã—ããŠè©±ã—ã™ã‚‹ã«ã‚ƒã€‚"
            },
            {
                "role": "user",
                "content": user_text
            }
        ]
    )

    # ãƒ†ã‚­ã‚¹ãƒˆå›ç­”ã‚’å–ã‚Šå‡ºã™ã«ã‚ƒ
    nyanko_text = completion.choices[0].message.audio.transcript
    print("\nã«ã‚ƒã‚“ã“ã®ãŠã¸ã‚“ã˜ã«ã‚ƒï¼š")
    print(nyanko_text)

    # éŸ³å£°ãƒ‡ãƒ¼ã‚¿ã‚‚å–ã‚Šå‡ºã—ã¦å†ç”Ÿã™ã‚‹ã«ã‚ƒ
    wav_bytes = base64.b64decode(completion.choices[0].message.audio.data)
    audio = Audio(wav_bytes)
    display(audio)

# ã«ã‚ƒã‚“ã“ãƒãƒ£ãƒƒãƒˆã‚’ã‚¹ã‚¿ãƒ¼ãƒˆã™ã‚‹ã«ã‚ƒï¼
nyanko_chat()
```

---

### ğŸ˜º ã‚¹ãƒ†ãƒƒãƒ—â‘¢ï¼šè³ªå•ã—ã¦ã¿ã‚‹ã«ã‚ƒï¼

ãŸã¨ãˆã°ã€ã“ã‚“ãªè³ªå•ã‚’ã—ã¦ã¿ã¦ã«ã‚ƒï¼š

- ã€Œé›¨ã®æ—¥ã£ã¦ã©ã†ã™ã”ã—ãŸã‚‰ã„ã„ã«ã‚ƒï¼Ÿã€
- ã€ŒãŠã™ã™ã‚ã®çŒ«ã®ãŠã‚‚ã¡ã‚ƒã‚ã‚‹ã«ã‚ƒï¼Ÿã€

---

### ğŸ“ æ³¨æ„ç‚¹ã ã«ã‚ƒï¼

- APIã‚­ãƒ¼ã¯çµ¶å¯¾ã«ä»–äººã¨å…±æœ‰ã—ãªã„ã§ã«ã‚ƒï¼
- ã“ã®ã‚³ãƒ¼ãƒ‰ã¯ `gpt-4o-audio-preview` ãƒ¢ãƒ‡ãƒ«ã‚’ä½¿ã£ã¦ã‚‹ã‹ã‚‰ã€**å¯¾å¿œã—ã¦ã‚‹ãƒ—ãƒ©ãƒ³ãŒå¿…è¦ã«ã‚ƒ**ã€‚
- éŸ³å£°ãƒ‡ãƒ¼ã‚¿ã¯ `.wav` å½¢å¼ã§è¿”ã£ã¦ãã‚‹ã«ã‚ƒã€‚Colabã§ãã®ã¾ã¾å†ç”Ÿã§ãã‚‹ã«ã‚ƒã€‚

````

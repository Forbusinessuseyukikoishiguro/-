````python
#🐾 Colabにゃんこチャットボット（GPT-4o音声付き）20250429

#「にゃんこチャット（音声対応🐾）」Colabでの使い方

---

## 🐱 にゃんこチャットの使い方（Google Colab版）

このにゃんこチャットでは、**テキストで質問を入力すると、かわいいにゃんこが声と文字でお返事してくれるにゃ！**
以下の手順でカンタンに使えるにゃ！

---

### 🪄 ステップ①：OpenAI APIキーを用意するにゃ！

[https://platform.openai.com/account/api-keys](https://platform.openai.com/account/api-keys)
上記リンクから **自分のAPIキー**（`sk-xxxxxxxxxxxxxxxxx`の形式）をコピーしておくにゃ！

---

### 🐾 ステップ②：Colabでコードを実行するにゃ！

1. Google Colabを開くにゃ。
2. 以下のコードを **コピペ**して、1セルずつ実行するにゃ。
3. 最初のほうで `APIキーを入力してください。` という場所があるから、自分のAPIキーに書き換えるにゃ！

```python
# 必要なライブラリをインストールするにゃ！
!pip install openai

# openaiライブラリを読み込むにゃ
import os
import base64
from openai import OpenAI
from IPython.display import Audio, display

# APIキーをここにセットするにゃ！（必ずご自身のキーに書き換えてにゃ！）
os.environ["OPENAI_API_KEY"] = "sk-xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx"

# OpenAIクライアントを作るにゃ
client = OpenAI()

# 🐾 にゃんこチャットスタート関数だにゃ！
def nyanko_chat():
    user_text = input("にゃんこに聞きたいことを入力してにゃ: ")

    completion = client.chat.completions.create(
        model="gpt-4o-audio-preview",
        modalities=["text", "audio"],
        audio={"voice": "nova", "format": "wav"},
        messages=[
            {
                "role": "system",
                "content": "あなたはかわいいにゃんこです。語尾に「にゃ」をつけて優しくお話しするにゃ。"
            },
            {
                "role": "user",
                "content": user_text
            }
        ]
    )

    # テキスト回答を取り出すにゃ
    nyanko_text = completion.choices[0].message.audio.transcript
    print("\nにゃんこのおへんじにゃ：")
    print(nyanko_text)

    # 音声データも取り出して再生するにゃ
    wav_bytes = base64.b64decode(completion.choices[0].message.audio.data)
    audio = Audio(wav_bytes)
    display(audio)

# にゃんこチャットをスタートするにゃ！
nyanko_chat()
```

---

### 😺 ステップ③：質問してみるにゃ！

たとえば、こんな質問をしてみてにゃ：

- 「雨の日ってどうすごしたらいいにゃ？」
- 「おすすめの猫のおもちゃあるにゃ？」

---

### 📝 注意点だにゃ！

- APIキーは絶対に他人と共有しないでにゃ！
- このコードは `gpt-4o-audio-preview` モデルを使ってるから、**対応してるプランが必要にゃ**。
- 音声データは `.wav` 形式で返ってくるにゃ。Colabでそのまま再生できるにゃ。

````
